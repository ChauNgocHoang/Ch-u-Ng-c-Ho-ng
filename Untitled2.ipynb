```python
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(6)
y = x + 1
plt.plot(x,y,'o--')
plt.show
```

```python
plt.plot(x,y,'o--')
plt.plot(x,-y,'o-')
plt.title('y = x & y = - x')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
x= np.arange(3)
plt.subplots_adjust(wspace=0.3,hspace=0.3)
plt.subplot(2,2,1)
plt.plot(x,-x**2,color='g')
plt.subplot(2,2,2)
plt.plot(x,-2*x,color='b')
plt.subplot(2,2,3)
plt.plot(x,-x**3,color='r')
plt.subplot(2,2,4)
plt.plot(x,-2**x,color='y')
plt.show()
```
